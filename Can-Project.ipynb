Create A Model:

```python
X=df.drop('benign_0__mal_1',axis=1).values
y=df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model=Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

#dinary Classification
model.add(Dense(1,activation='sigmoid'))


model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,validation_data=(X_test,y_test))
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses
```

```python
plt.figure(figsize=(10,6))
plt.plot(losses)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
earlystop=EarlyStopping(mode='min',monitor='val_loss',verbose=1,patience=15)
```

```python
model=Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu'))

#dinary Classification
model.add(Dense(1,activation='sigmoid'))


model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),epochs=600,callbacks=[earlystop])
```

```python
mod_loss=pd.DataFrame(model.history.history)
```

```python
plt.plot(mod_loss)
```

```python
from tensorflow.keras.layers import Dropout
```

```python
model=Sequential()
model.add(Dense(30,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))

#dinary Classification
model.add(Dense(1,activation='sigmoid'))


model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),epochs=600,callbacks=[earlystop])
```

```python
mod_losses=pd.DataFrame(model.history.history)
```

```python
plt.plot(mod_losses)
```

```python
predictions=(model.predict(X_test)>0.5).astype('int32')
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python

print(classification_report(predictions, y_test))
```

```python
print(confusion_matrix(predictions,y_test))
```
